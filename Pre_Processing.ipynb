```python
import os
import pydicom as dicom
from PIL import Image

# Input and output directories
input_dir = "C:\\Users\\chinnalab123\\Downloads\\INbreast Release 1.0\\AllDICOMs\\"  # Update with the path to your input folder containing DICOM files
output_dir = "C:\\Users\\chinnalab123\\Downloads\\INbreast Release 1.0\\resized_png\\"  # Update with the path to your output folder for PNG files

# Ensure the output directory exists, create if not
os.makedirs(output_dir, exist_ok=True)

# Iterate through all files in the input directory
for filename in os.listdir(input_dir):
    if filename.endswith(".dcm"):  # Process only DICOM files
        # Read the DICOM file
        ds = dicom.dcmread(os.path.join(input_dir, filename))
        fname=filename.strip(".dcm")
#         print(fname)
        # Extract pixel data from the DICOM file
        pixel_data = ds.pixel_array

        # Resize the image to 224x224
        resized_image = Image.fromarray(pixel_data).resize((224, 224))

        # Save the resized image as PNG in the output directory with the same filename

        #resized_image.save(output_dir+fname+".png")
        plt.imsave(output_dir+fname+".png", resized_image, cmap='gray')
        plt.axis('off')


print("Conversion completed.")
```

```python
import os
import pandas as pd
import numpy as np
import shutil

# Load the CSV file
csv_file_path = 'C:\\Users\\chinnalab123\\Downloads\\INbreast Release 1.0\\INbreast.csv'

# Read the CSV file into a DataFrame
df = pd.read_csv(csv_file_path, sep=';')

# Extract the filenames and Bi-Rads labels as NumPy arrays
filenames = df.iloc[:, 5].values
birad_labels = df.iloc[:, -1].values

# Convert all values other than '1' to '0' in the Bi-Rads labels
birad_labels = np.where(birad_labels == '1', 1, 0)

# Create a dictionary to map filenames to Bi-Rads labels
filename_to_birad = dict(zip(filenames, birad_labels))

# Define the input and output directories
input_dir = 'C:\\Users\\chinnalab123\\Downloads\\INbreast Release 1.0\\resized_png\\'  # Update with the path to your input directory
normal_folder = 'C:\\Users\\chinnalab123\\Downloads\\INbreast Release 1.0\\resized_png\\normal\\'  # Update with the path to your normal folder

# Ensure the normal folder exists, create if not
os.makedirs(normal_folder, exist_ok=True)

# Move files starting with elements in ab_norm to the normal folder
for filename in os.listdir(input_dir):
    if filename.startswith(tuple(ab_norm.astype('str'))):
        # Get the Bi-Rads label for the filename
        birad_label = filename_to_birad.get(filename, 0)

        # Move the file to the normal folder
        shutil.move(os.path.join(input_dir, filename), os.path.join(normal_folder, filename))

print("Files moved to the normal folder.")
```
